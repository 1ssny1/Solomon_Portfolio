This case study analyzes data from the Government of Canada's Open Government Portal, with a specific focus on contract awards. The analysis utilizes data from the 2024-2025 fiscal year, representing the most recent complete dataset available. The case study is structured into three distinct parts, as outlined below:
1. Data acquisition and preparation which will deal with downloading the data and cleaning it for further analysis.
2.	Audit Objectives & Hypotheses section which deals with creating and testing different hypotheses regarding the contract award process.
3.	Data Analysis & Insights section where PowerBi will be used to identify and describe valuable audit or compliance-related insights.



## Section 1. Data Acquisition and preparation.

Data for the 2024-2025 fiscal year was downloaded from the Government of Canada's Open Government Portal, as a CSV file and loaded into a pandas’ data frame. Python will be used for this section.


```python
# Importing Functions
import pandas as pd
import numpy as np

```

```python
# Load the dataframe.
df_award_notice = pd.read_csv('2024-2025-awardNotice-avisAttribution.csv')

# Display the first 5 rows.
display(df_award_notice.head())
df_award_notice.info()
df_award_notice.describe()
```


Initial analysis of the dataset reveals the following:
1. 	There is a total of 76 columns and 6571 rows. The columns have the names in both English and French with a dash in between.
2. 	Only 3 columns have a numeric datatype, the rest have an object datatype including the date columns
  



### Data cleanup

The column titles are edited to remove the French names, and the duplicated columns are deleted. This leaves a total of 49 unique columns.


```python
# Remove french names from the columns
df_award_notice.columns = [col.split('-')[0] for col in df_award_notice.columns]

# Find and remove duplicated columns
df_award_notice = df_award_notice.loc[:,~df_award_notice.columns.duplicated()]
display(df_award_notice.head())
```

#### Date Columns

##### Several columns, including publicationDate, contractAwardDate, amendmentDate, contractStartDate, and contractEndDate, were initially of object type and are converted to datetime objects.


```python
# 2. Convert date columns to datetime objects
date_cols = ['publicationDate', 'contractAwardDate', 'amendmentDate', 'contractStartDate', 'contractEndDate']

for col in date_cols:
    df_award_notice[col] = pd.to_datetime(df_award_notice[col], errors='coerce')

print("Number of NaT values introduced after converting to datetime:")
for col in date_cols:
    # Re-calculate missing_values since the dataframe changed
    missing_values_after_drop = df_award_notice.isnull().sum()
    nat_count = df_award_notice[col].isnull().sum() - missing_values_after_drop[col]
    print(f"- {col}: {nat_count} NaT values")
```

#### Inconsistent Data

Inconsistencies were found and addressed in the following categorical columns: procurementCategory, supplierAddressCountry, contractingEntityAddressProvince, contractingEntityAddressCountry, and contactInfoCountry. These inconsistencies included variations in spelling, capitalization, extra spaces, and non-standard entries.


Analyse categorical columns

```python
# Analyse categorical columns
categorical_cols = df_award_notice.select_dtypes(include='object').columns

print("Number of unique values for each categorical column:")
for col in categorical_cols:
    num_unique = df_award_notice[col].nunique()
    print(f"- {col}: {num_unique} unique values")
    if num_unique < 50:
        print(f"  Unique values for {col}:")
        display(df_award_notice[col].unique())
```

#### Missing Values

The columns with missing values were analysed and changes were made as follows:
1. gsin ( Goods and Services Identification Number) column has 96% of the data missing and was deleted as this is a legacy code no longer used according to the official documentation. This is the same for the gsinDescription column as well. 
2.	A total of 7 columns containing contact info for tendering organisations or individuals had more than 70% of the data missing and these were deleted given the context. The columns are contactInfoAddressLine, contactInfoPostalcode, contactInfoProvince, contactInfoCity, contactInfoCountry, contactInfoPhone and contactInfoFax. Similarly named columns with supplier contact information had insignificant missing values indicating that these 7 columns are most likely duplicates.
3.	Some entries for contractAmount and totalContractValue had negative values and given the context of contract awards, these were most likely errors and were set to the absolute figure. Missing values in these 2 columns were imputed to 0.
4.	Non-datetime, non-numeric columns with missing values between 10% and 50%, missing values were imputed with the string 'Unknown'. For categorical columns in this range, 'Unknown' was added as a category before imputation.


```python
# Show percentage of missing values per column
missing_values = df_award_notice.isnull().sum()
missing_percentage = (missing_values / len(df_award_notice)) * 100

missing_info = pd.DataFrame({'Missing Count': missing_values, 'Missing Percentage (%)': missing_percentage})
missing_info = missing_info[missing_info['Missing Count'] > 0].sort_values(by='Missing Count', ascending=False)

print("Missing values per column:")
display(missing_info)
```

```python

#  Drop columns with very high percentage of missing values 
cols_to_drop = [
    'gsin', 'gsinDescription', 'contactInfoAddressLine', 'contactInfoPostalcode',
    'contactInfoProvince', 'contactInfoCity', 'contactInfoCountry', 'contactInfoPhone',
    'contactInfoFax'
]
cols_to_drop_existing = [col for col in cols_to_drop if col in df_award_notice.columns]

if cols_to_drop_existing:
    df_award_notice = df_award_notice.drop(columns=cols_to_drop_existing)
    print(f"Columns dropped: {cols_to_drop_existing}")
else:
    print("No specified columns with high missing values to drop.")

print(f"Number of columns after dropping: {df_award_notice.shape[1]}")

# Convert date columns to datetime objects (already done in previous step, but included for completeness)
date_cols = ['publicationDate', 'contractAwardDate', 'amendmentDate', 'contractStartDate', 'contractEndDate']
for col in date_cols:
    if col in df_award_notice.columns: # Check if column exists after dropping
        df_award_notice[col] = pd.to_datetime(df_award_notice[col], errors='coerce')


#  Address inconsistencies in categorical columns
# Standardize country names
country_cols = ['supplierAddressCountry', 'contractingEntityAddressCountry', 'contactInfoCountry']
for col in country_cols:
    if col in df_award_notice.columns: # Check if column exists after dropping
        df_award_notice[col] = df_award_notice[col].astype(str).str.strip().str.upper().replace({
            'CA': 'CANADA', 'CANDA': 'CANADA', 'CAN': 'CANADA', '': None, 'NAN': None
        })


# Standardize province names
province_cols = ['contractingEntityAddressProvince']
for col in province_cols:
     if col in df_award_notice.columns: # Check if column exists after dropping
        df_award_notice[col] = df_award_notice[col].astype(str).str.strip().replace({
            'QC': 'Quebec', 'Québec': 'Quebec', 'ON': 'Ontario', 'On': 'Ontario',
            'National Capital Region (NCR)': 'Ontario', 'Ottawa': 'Ontario',
            'Ontarion': 'Ontario', 'nan': None
        })


# Clean procurement category variations
if 'procurementCategory' in df_award_notice.columns: # Check if column exists after dropping
    df_award_notice['procurementCategory'] = df_award_notice['procurementCategory'].astype(str).replace({
        '\n': '', 'SRVTGD': 'SRV*GD'
    }, regex=True).str.replace(r'\*GD\*SRV', '*SRV*GD').replace('nan', None)


#  Handle missing values 
# Address negative 'contractAmount' and 'totalContractValue' by setting them to their absolute value
if 'contractAmount' in df_award_notice.columns:
    df_award_notice['contractAmount'] = df_award_notice['contractAmount'].abs()
if 'totalContractValue' in df_award_notice.columns:
    df_award_notice['totalContractValue'] = df_award_notice['totalContractValue'].abs()

# Impute missing 'contractAmount' and 'totalContractValue' with 0
if 'contractAmount' in df_award_notice.columns:
    df_award_notice['contractAmount'] = df_award_notice['contractAmount'].fillna(0)
if 'totalContractValue' in df_award_notice.columns:
    df_award_notice['totalContractValue'] = df_award_notice['totalContractValue'].fillna(0)

# Impute missing values in non-datetime, non-numeric columns (with missing values between 10% and 50%) with 'Unknown'
# Identify non-datetime, non-numeric columns
object_cols = df_award_notice.select_dtypes(include='object').columns
datetime_cols_existing = [col for col in date_cols if col in df_award_notice.columns]
numerical_cols_existing = [col for col in ['amendmentNumber', 'contractAmount', 'totalContractValue'] if col in df_award_notice.columns]

cols_to_impute_unknown = [col for col in object_cols if col not in datetime_cols_existing and col not in numerical_cols_existing]

# Check missing percentage for these columns and impute if between 10% and 50%
for col in cols_to_impute_unknown:
    missing_percentage = (df_award_notice[col].isnull().sum() / len(df_award_notice)) * 100
    if 10 <= missing_percentage <= 50:
         # Add 'Unknown' as a category if it's a categorical column
        if pd.api.types.isinstance(dtype, pd.CategoricalDtype)(df_award_notice[col]):
            if 'Unknown' not in df_award_notice[col].cat.categories:
                df_award_notice[col] = df_award_notice[col].cat.add_categories('Unknown')
        df_award_notice[col] = df_award_notice[col].fillna('Unknown')
        print(f"Imputed missing values in column '{col}' with 'Unknown'.")

# 5. Address potential outliers in numerical columns (as before, no removal/capping without further investigation)

# Display descriptive statistics after cleaning
display(df_award_notice.describe(include='all'))
```

#### Data Transformation

The difference between contractEndDate and contractStartDate was calculated and stored in the contractDuration column.

The following columns were converted to category datatype for better analysis: contractCurrency, awardStatus, procurementCategory, noticeType, procurementMethod, selectionCriteria, limitedTenderingReason, contractingEntityAddressProvince, contractingEntityAddressCountry and     contactInfoCountry.

The text fields (title and awardDescription) were standardized by converting them to lowercase, removing leading and trailing whitespace, and replacing 'nan' strings with None.


```python
# Calculate contract duration column
df_award_notice['contractDuration'] = df_award_notice['contractEndDate'] - df_award_notice['contractStartDate']

# Convert some columns to categorical datatypes
categorical_cols = ['contractCurrency', 'awardStatus', 'procurementCategory', 'noticeType', 'procurementMethod', 'selectionCriteria', 'limitedTenderingReason', 'contractingEntityAddressProvince', 'contractingEntityAddressCountry', 'contactInfoCountry']

for col in categorical_cols:
    if col in df_award_notice.columns:
        df_award_notice[col] = df_award_notice[col].astype('category')

#Standardize the 'title' and 'awardDescription' columns
cols_to_standardize = ['title', 'awardDescription']

for col in cols_to_standardize:
    if col in df_award_notice.columns:
        df_award_notice[col] = df_award_notice[col].astype(str).str.lower().str.strip().replace('nan', None)

display(df_award_notice.info())
display(df_award_notice.head())
```
